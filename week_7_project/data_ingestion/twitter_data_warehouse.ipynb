```python
import pandas as pd
```

```python
_2010 = pd.read_csv('../data_ingestion/data/twitter/csv/2010.csv')
_2011 = pd.read_csv('../data_ingestion/data/twitter/csv/2011.csv')
_2012 = pd.read_csv('../data_ingestion/data/twitter/csv/2012.csv')
_2013 = pd.read_csv('../data_ingestion/data/twitter/csv/2013.csv')
_2014 = pd.read_csv('../data_ingestion/data/twitter/csv/2014.csv')
_2015 = pd.read_csv('../data_ingestion/data/twitter/csv/2015.csv')
_2016 = pd.read_csv('../data_ingestion/data/twitter/csv/2016.csv')
_2017 = pd.read_csv('../data_ingestion/data/twitter/csv/2017.csv')
_2018 = pd.read_csv('../data_ingestion/data/twitter/csv/2018.csv')
_2019 = pd.read_csv('../data_ingestion/data/twitter/csv/2019.csv')
_2020 = pd.read_csv('../data_ingestion/data/twitter/csv/2020.csv')
_2021 = pd.read_csv('../data_ingestion/data/twitter/csv/2021.csv')
```

```python
tweets = pd.concat([_2010, _2011, _2012, _2013, _2014, _2015, _2016, _2017, _2018, _2019, _2020, _2021], axis = 0)
```

```python
tweets.head()
```

```python
tweets = tweets.reset_index(drop=True)
tweets
```

```python
tweets.info()
```

```python
tweets.describe()
```

```python
tweets = tweets[['date','replies_count', 'retweets_count', 'likes_count']]
tweets
```

```python
#Lets clean up the date format
tweets.date = tweets.date.str[:10] #tweets.date.str[:10]
tweets
```
